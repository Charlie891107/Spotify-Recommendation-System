##Model 1: Cosine Similarity
This is an unsupervised method where you recommend items (albums) based on their similarity to a user's preferences (derived from their top 5 songs). The idea is to recommend albums whose song attributes closely match the aggregated features of the user's top songs.
Cosine Similarity can be used to measure the similarity between the user's profile and the album's features.



```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

def recommend_album_nameviz(user_top_songs_file, normalized_df, scaler):
  """
    Recommends an album based on the user's top songs,
    and prints a histogram to visualize the similarity scores
    """
  # Aggregating the song features by album (using mean)
  normalized_album_features = normalized_df.groupby('album_name').mean()

  # Assume user has top 5 songs
  user_songs_df = pd.read_csv(user_top_songs_file)
  #remove 'song' column and normalize data
  user_songs_df = user_songs_df.drop('Song', axis=1)
  user_songs_df = scaler.transform(user_songs_df)

  # Calculate the user's profile by averaging their top songs' features
  user_profile = np.mean(user_songs_df, axis=0)

  # Now compute cosine similarity between user profile and album features
  album_similarity = cosine_similarity(user_profile.reshape(1, -1), normalized_album_features)

  # Get the most similar album
  most_similar_album_idx = np.argmax(album_similarity)
  most_similar_album = normalized_album_features.index[most_similar_album_idx]

  print(f"Recommended Album: {most_similar_album}")
  # Reshape cosine similarity scores for visualization
  similarity_scores = album_similarity.flatten()

  # Create a DataFrame for better handling
  similarity_df = pd.DataFrame({
      'Album': normalized_album_features.index,
      'Similarity': similarity_scores
  }).sort_values(by='Similarity', ascending=False)

  # Bar chart for similarity scores
  plt.figure(figsize=(12, 6))
  sns.barplot(
      x='Similarity',
      y='Album',
      data=similarity_df,
      hue='Album',
      palette="viridis"
  )
  plt.title('Cosine Similarity Scores for Albums')
  plt.xlabel('Similarity Score')
  plt.ylabel('Album')
  plt.axvline(x=similarity_df['Similarity'].max(), color='red', linestyle='--')
  plt.tight_layout()
  plt.show()
```

##Model 2: Clustering
By clustering songs with similar features, you can assign a user to the cluster that best represents their music taste.

We then determine which album to recommend by finding the album with the highest number of recommended songs

###Determining Number of Clusters

####Elbow Test
This shows that the 3 is the optimal number of clusters to capture differernces in attributes between songs

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load and preprocess data (assuming normalized_df is already prepared)
# Here normalized_features represents the normalized feature data for clustering
# Load the dataset
taylor_df = pd.read_csv('taylor_swift_songs.csv')

# Normalize the features using MinMaxScaler
features = taylor_df[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'tempo']]
scaler = MinMaxScaler()
scaler.fit(features)
normalized_features = scaler.transform(features)

# Combine the normalized features and album/track info
normalized_df = pd.DataFrame(normalized_features, columns=features.columns)
normalized_df['album_name'] = taylor_df['album_name']
normalized_df['track_name'] = taylor_df['track_name']
#print(normalized_df.head())


# Calculate inertia for different numbers of clusters
inertia = []
cluster_range = range(1, 15)  # Test with 1 to 14 clusters

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=5)
    kmeans.fit(normalized_df.drop(['track_name', 'album_name'], axis=1))
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, inertia, marker='o', color='b', linestyle='--')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```

####3 Clusters
We see with 3 clusters however, that many albums may be recommended, so we test out different numbers of clusters.

```python
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

#variables to change
n_clusters = 3
r = 100 #random state
user_top_songs_file = "Riya_top_songs.csv"

# Load the dataset
taylor_df = pd.read_csv('taylor_swift_songs.csv')

# Normalize the features using MinMaxScaler
features = taylor_df[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'tempo']]
scaler = MinMaxScaler()
scaler.fit(features)
normalized_features = scaler.transform(features)

# Combine the normalized features and album/track info
normalized_df = pd.DataFrame(normalized_features, columns=features.columns)
normalized_df['album_name'] = taylor_df['album_name']
normalized_df['track_name'] = taylor_df['track_name']
#print(normalized_df.head())

# 1. Perform K-Means clustering to group songs into clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=r)
normalized_df['cluster'] = kmeans.fit_predict(normalized_df.drop(['track_name', 'album_name'], axis=1))


# Calculate this user profile from their top songs
user_songs_df = pd.read_csv(user_top_songs_file)
user_songs_df = user_songs_df.drop('Song', axis=1)
user_songs_df = scaler.transform(user_songs_df)
user_profile = np.mean(user_songs_df, axis=0).reshape(1, -1)

# 3. Match the user profile with each cluster's center using cosine similarity
cluster_centers = kmeans.cluster_centers_
cluster_similarity = cosine_similarity(user_profile, cluster_centers)

# Identify the most similar cluster to the user profile
most_similar_cluster_idx = np.argmax(cluster_similarity)

# 4. Filter songs that belong to the most similar cluster
cluster_songs_df = normalized_df[normalized_df['cluster'] == most_similar_cluster_idx]

# 5. Return the song names and album names from that cluster
recommended_songs = cluster_songs_df[['track_name', 'album_name']]

# 6. Find the album with the most recommended songs in the most similar cluster
album_counts = cluster_songs_df['album_name'].value_counts()

# The most recommended album will be the one with the highest count of songs in the cluster
top_album = album_counts.idxmax()

#print results
print(f"Cluster Number: {most_similar_cluster_idx}")
print(f"Top Recommended Album: {top_album}")
print("Songs in the most similar cluster:")
print(recommended_songs)

print("\nRecommended Albums for Each Cluster:")
for cluster_idx in range(n_clusters):
    cluster_songs_df = normalized_df[normalized_df['cluster'] == cluster_idx]
    album_counts = cluster_songs_df['album_name'].value_counts()
    top_album = album_counts.idxmax()  # Get the album with the most songs in this cluster
    print(f"Cluster {cluster_idx}: {top_album}")

#Visualize the cluster distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=normalized_df, x='cluster', hue ='cluster', palette='viridis' )
plt.title('Distribution of Songs across Clusters')
plt.xlabel('Cluster')
plt.ylabel('Number of Songs')
plt.legend().remove()
plt.show()

#Visualize the clusters, centers, and user profile location
# Step 1: Dimensionality reduction to 2D (using PCA for visualization purposes)
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(normalized_features)
user_profile_pca = pca.transform(user_profile)

# Step 2: Plotting the clusters and user profile
plt.figure(figsize=(15, 12))

# Plot the songs (colored by their cluster)
sns.scatterplot(x=reduced_features[:, 0], y=reduced_features[:, 1], hue=normalized_df['cluster'], palette='tab10', s=100, alpha=0.6)

# Plot the cluster centers (the centroids of the clusters)
centroids_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c='black', marker='x', s=200, label='Cluster Centers')

# Plot the user profile location (as a red star)
plt.scatter(user_profile_pca[:, 0], user_profile_pca[:, 1], c='red', marker='*', s=300, label='User Profile')

# Add labels and legend
plt.title(f"{n_clusters} Clusters: Song Clusters, User Profile, and Cluster Centers")
plt.legend()

# Show the plot
plt.show()

# Filter the songs in the cluster matched to the user profile
cluster_songs_df = normalized_df[normalized_df['cluster'] == most_similar_cluster_idx]

# Count the number of songs per album in this cluster
album_counts = cluster_songs_df['album_name'].value_counts()

# Create a bar plot for the album counts
plt.figure(figsize=(12, 6))
sns.barplot(x=album_counts.index, y=album_counts.values, palette='coolwarm')

# Customize the plot
plt.title(f"Album Song Counts in Cluster {most_similar_cluster_idx}")
plt.xlabel('Album Name')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45, ha='right')  # Rotate album names for better readability

# Show the plot
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

#variables to change
n_clusters = 10
r = 100 #random state
user_top_songs_file = "Claire_top_songs.csv"

# Load the dataset
taylor_df = pd.read_csv('taylor_swift_songs.csv')

# Normalize the features using MinMaxScaler
features = taylor_df[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'tempo']]
scaler = MinMaxScaler()
scaler.fit(features)
normalized_features = scaler.transform(features)

# Combine the normalized features and album/track info
normalized_df = pd.DataFrame(normalized_features, columns=features.columns)
normalized_df['album_name'] = taylor_df['album_name']
normalized_df['track_name'] = taylor_df['track_name']
#print(normalized_df.head())

# 1. Perform K-Means clustering to group songs into clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=r)
normalized_df['cluster'] = kmeans.fit_predict(normalized_df.drop(['track_name', 'album_name'], axis=1))


# Calculate this user profile from their top songs
user_songs_df = pd.read_csv(user_top_songs_file)
user_songs_df = user_songs_df.drop('Song', axis=1)
user_songs_df = scaler.transform(user_songs_df)
user_profile = np.mean(user_songs_df, axis=0).reshape(1, -1)

# 3. Match the user profile with each cluster's center using cosine similarity
cluster_centers = kmeans.cluster_centers_
cluster_similarity = cosine_similarity(user_profile, cluster_centers)

# Identify the most similar cluster to the user profile
most_similar_cluster_idx = np.argmax(cluster_similarity)

# 4. Filter songs that belong to the most similar cluster
cluster_songs_df = normalized_df[normalized_df['cluster'] == most_similar_cluster_idx]

# 5. Return the song names and album names from that cluster
recommended_songs = cluster_songs_df[['track_name', 'album_name']]

# 6. Find the album with the most recommended songs in the most similar cluster
album_counts = cluster_songs_df['album_name'].value_counts()

# The most recommended album will be the one with the highest count of songs in the cluster
top_album = album_counts.idxmax()

#print results
print(f"Cluster Number: {most_similar_cluster_idx}")
print(f"Top Recommended Album: {top_album}")
#print("Songs in the most similar cluster:")
#print(recommended_songs)

print("\nRecommended Albums for Each Cluster:")
for cluster_idx in range(n_clusters):
    cluster_songs_df = normalized_df[normalized_df['cluster'] == cluster_idx]
    album_counts = cluster_songs_df['album_name'].value_counts()
    top_album = album_counts.idxmax()  # Get the album with the most songs in this cluster
    print(f"Cluster {cluster_idx}: {top_album}")

#Visualize the cluster distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=normalized_df, x='cluster', hue ='cluster', palette='viridis' )
plt.title('Distribution of Songs across Clusters')
plt.xlabel('Cluster')
plt.ylabel('Number of Songs')
plt.legend().remove()
plt.show()

#Visualize the clusters, centers, and user profile location
# Step 1: Dimensionality reduction to 2D (using PCA for visualization purposes)
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(normalized_features)
user_profile_pca = pca.transform(user_profile)

# Step 2: Plotting the clusters and user profile
plt.figure(figsize=(15, 12))

# Plot the songs (colored by their cluster)
sns.scatterplot(x=reduced_features[:, 0], y=reduced_features[:, 1], hue=normalized_df['cluster'], palette='tab10', s=100, alpha=0.6)

# Plot the cluster centers (the centroids of the clusters)
centroids_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c='black', marker='x', s=200, label='Cluster Centers')

# Plot the user profile location (as a red star)
plt.scatter(user_profile_pca[:, 0], user_profile_pca[:, 1], c='red', marker='*', s=300, label='User Profile')

# Add labels and legend
plt.title(f"{n_clusters} Clusters: Song Clusters, User Profile, and Cluster Centers")
plt.legend()

# Show the plot
plt.show()

# Filter the songs in the cluster matched to the user profile
cluster_songs_df = normalized_df[normalized_df['cluster'] == most_similar_cluster_idx]

# Count the number of songs per album in this cluster
album_counts = cluster_songs_df['album_name'].value_counts()

# Create a bar plot for the album counts
plt.figure(figsize=(12, 6))
sns.barplot(x=album_counts.index, y=album_counts.values, palette='coolwarm')

# Customize the plot
plt.title(f"Album Song Counts in Cluster {most_similar_cluster_idx}")
plt.xlabel('Album Name')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45, ha='right')  # Rotate album names for better readability

# Show the plot
plt.show()
```
